1. Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

___MEAN (RATA RATA)___

```python

# inisialisasi dataset
data = [100.5 , 99.7 , 101 , 101.3 , 99.9 , 98.9 , 165 , 102.2 , 98.7 , 102.9 , 100]
print(f"Data: {data}")

# mencari penjumlahan nilai data
jumlah_nilai = sum(data)
print(f"Jumlah nilai data: {jumlah_nilai}")

# mencari total data dalam datasaet
total_data = len(data)
print(f"Total data: {total_data}")

# menghitung mean
mean_data = jumlah_nilai / total_data
print(f"Mean data: {mean_data}")

```

Jika divisualisasikan dalam bentuk grafik, maka mean data tersebut akan terlihat seperti ini:

```python
# Import library
from matplotlib import pyplot as plt
# Buat histogram
plt.hist(data, bins= 4, edgecolor='black', color='lightskyblue')

# Menambahkan sebuah garis vertikal pada titik mean
plt.axvline(x=mean_data, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_data:.2f}')

# Menambahkan judul dan label
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Mean')

# Menambahkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

Dari visualisasi data tersebut, dapat terlihat bahwa terdapat data yang sangat berbeda atau "__Outlier__" yang memengaruhi data, sehingga mean menjadi kurang mewakili pusat data yang sebenarnya. 

___MEDIAN___

```python
#Mengurutkan data
data_urut = sorted(data)
print(f"Data berurut: {data_urut}")
print(f"Total data: {total_data}")

nilai_tengah = total_data//2 # // untuk floor division, atau pembagian dengan pembulatan ke bawah

# Karena jumlah data ganjil, maka nilai tengah bisa didapatkan dengan cara mengambil nilai tengah
print(f"Jumlah total data ganjil")
median_data = data_urut[nilai_tengah]
print(f"Nilai tengah: {median_data}")
```

Visualisasi dalam bentuk grafik:

```python
# Import library
from matplotlib import pyplot as plt

# buat histogram
plt.hist(data, bins= 4, edgecolor='black', color='lightskyblue')

# Menambahkan sebuah garis vertikal pada titik mean
plt.axvline(x=median_data, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_data}')

# Menambahkan judul dan label
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Median')

# Menambahkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

Pada penggambaran median, persebaran data lebih jelas

___MODUS (MODE)___

```python
# Import library Counter untuk menghitung modus
from collections import Counter

print(f"Data: {data}")

counter = Counter(data)
modus_data = dict(counter) # ubah ke tipe data dictionary
max_counter = max(list(counter.values())) # mencari data yang memiliki counter paling banyak
print(f"Jumlah kemunculan tertinggi: {max_counter}")

nilai_modus = []
for k, v in modus_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        nilai_modus.append(k)
    
print(f"Modus: {nilai_modus}")
```

Berdasarkan hasil tersebut, dapat disimpulkan bahwa data tersebut tidak memiliki modus, dikarenakan semua nilai pada tersebut memiliki nilai yang berbeda, visualisasinya akan berberntuk seperti ini:


```python
# Import library
from matplotlib import pyplot as plt

# buat histogram
plt.hist(data, bins= 4, edgecolor='black', color='lightskyblue')

# Menambahkan judul dan label
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Modus')

# Menampilkan histogram
plt.show()
```

___NOMOR 2___

2. ___Berikut adalah nilai UAS pada mata pelajaran matematika di kelas A: 95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?___


___MEAN (RATA RATA)___

```python
#deklarasi
data2 = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]
print(f"Data: {data2}")

# mencari penjumlahan nilai data
jumlah_nilai2= sum(data2)
print(f"Jumlah nilai data: {jumlah_nilai2}")

# mencari total data dalam datasaet
total_data2 = len(data2)
print(f"Total data: {total_data2}")

# menghitung mean
mean_data2 = jumlah_nilai2 / total_data2
print(f"Mean data: {mean_data2}")
```

Visualisasi dalam grafik:

```python
# Import library
from matplotlib import pyplot as plt

# buat histogram
plt.hist(data2, bins= 6, edgecolor='black', color='lightskyblue')

# Menambahkan sebuah garis vertikal pada titik mean
plt.axvline(x=mean_data2, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_data2:.2f}')

# Menambahkan judul dan label
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Mean')

# Menambahkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

Berdasarkan visualisasi tersbeut, terlihat bahwa persebaran data cukup merata, di mana nilai rata rata berada di mendekati tengah tengah

___MEDIAN___

```python
#Mengurutkan data
data_urut2 = sorted(data2)
print(f"Data berurut: {data_urut2}")
print(f"Total data: {total_data2}")

nilai_tengah2 = total_data2 // 2 # // untuk floor division, atau pembagian dengan pembulatan ke bawah
# Karena jumlah data genap, maka nilai tengah bisa didapatkan dengan cara
print(f"Jumlah total data genap")
median1 = data_urut2[nilai_tengah2]
median2 = data_urut2[nilai_tengah2 - 1]
print(f"Nilai tengah: {median1} {median2}")
median_data2 = (median1 + median2) / 2
print(f"Median: {median_data2}")
```

Hal tersebut dapat divisualisasikan menjadi:

```python
# Import library
from matplotlib import pyplot as plt

# buat histogram
plt.hist(data2, bins= 6, edgecolor='black', color='lightskyblue')

# Menambahkan sebuah garis vertikal pada titik mean
plt.axvline(x=median_data2, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_data2}')

# Menambahkan judul dan label
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Median')

# Menambahkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

Jika dilihat dan dibandingkan dengan visualisasi mean, dapat dilihat bahwa perbedaan dari median dan mean pada data ini tidakah jauh berbeda, hal ini menandakan bahwa persebaran data pada data ini cukup merata

___MODUS (MODE)___

```python
# Import library Counter untuk menghitung modus
from collections import Counter

print(f"Data: {data2}")

counter2 = Counter(data2)
modus_data2 = dict(counter2) # ubah ke tipe data dictionary
max_counter2 = max(list(counter2.values())) # mencari data yang memiliki counter paling banyak
print(f"Jumlah kemunculan tertinggi: {max_counter2}")

nilai_modus2 = []
for a, b in modus_data2.items():
    print(f"{a} muncul {b} kali")
    if b == max(list(counter2.values())):
        nilai_modus2.append(a)
    
print(f"Modus: {nilai_modus2}")
```

Hal ini dapat divisualisasikan menjadi:

```python
# Import library
from matplotlib import pyplot as plt

# Atur ukuran
plt.figure(figsize=(10,6))

# buat histogram
plt.hist(data2, bins= 6, edgecolor='black', color='lightskyblue')

# Add a vertical line at the mean
plt.axvline(x=nilai_modus2[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {nilai_modus2[0]:.2f}')
plt.axvline(x=nilai_modus2[1], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 2 = {nilai_modus2[1]:.2f}')
plt.axvline(x=nilai_modus2[2], color='green', linestyle='dashed', linewidth=1, label=f'Modus 1 = {nilai_modus2[2]:.2f}')
plt.axvline(x=nilai_modus2[3], color='gray', linestyle='dashed', linewidth=1, label=f'Modus 2 = {nilai_modus2[3]:.2f}')
plt.axvline(x=nilai_modus2[4], color='yellow', linestyle='dashed', linewidth=1, label=f'Modus 2 = {nilai_modus2[4]:.2f}')

# Menambahkan judul dan label
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Modus')

# Menampilkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

Berdasarkan visualisasi data tersebut, terlihat jelas bahwa data tersebut memiliki multimodal atau beberapa nilai modus.

___KESIMPULAN___:
Bila dibandingkan dengan data ke satu, data kedua memiliki persebaran data yang lebih merata, hal ini dapat dilihat terutama dari visualisasi nilai mean dan median di mana pada data ke satu mean lumayan melenceng dari nilai tenagh persebaran data dikarenakan data ke satu memiliki data __"Outlier"__.
